# Promedio M贸vil en Finanzas con Python y pandas

El **promedio m贸vil** (o *moving average*) es una t茅cnica que calcula el **promedio de un conjunto de datos dentro de una ventana de tiempo**, y esa ventana se va moviendo d铆a a d铆a.

 Sirve para **suavizar los precios** y observar mejor la **tendencia general** de un activo.

---

##  Ejemplo

Supongamos que tienes los precios de una acci贸n durante 5 d铆as:

| D铆a | Precio |
|----|------|
| 1  | 10   |
| 2  | 12   |
| 3  | 14   |
| 4  | 16   |
| 5  | 18   |

Si calculamos un **promedio m贸vil simple (SMA) de 3 d铆as**:

- D铆a 3  (10 + 12 + 14) / 3 = **12.0**  
- D铆a 4  (12 + 14 + 16) / 3 = **14.0**  
- D铆a 5  (14 + 16 + 18) / 3 = **16.0**  

 El promedio sigue la tendencia general, pero es m谩s *suave* que los precios diarios.

---

##  F贸rmulas en LaTeX

### 1. **Promedio M贸vil Simple (SMA)**  
El **SMA** de per铆odo \( n \) en el tiempo \( t \) se define como:

$$
SMA_t = \frac{1}{n} \sum_{i=0}^{n-1} P_{t-i}
$$

donde:  
- \( P_{t} \) = precio en el d铆a \( t \)  
- \( n \) = n煤mero de d铆as en la ventana  

---

### 2. **Promedio M贸vil Exponencial (EMA)**  
El **EMA** da m谩s peso a los datos recientes y se define como:

$$
EMA_t = \alpha \cdot P_t + (1 - \alpha) \cdot EMA_{t-1}
$$

donde:  
- \( P_t \) = precio en el d铆a \( t \)  
- \( EMA_{t-1} \) = EMA del d铆a anterior  
- \( \alpha = \frac{2}{n+1} \) = factor de suavizamiento  

---

##  驴Qu茅 es `.rolling()` en pandas?

El m茅todo **`.rolling(n)`** en pandas crea una **ventana de tama帽o `n`** que se va desplazando a lo largo de la serie de datos.  

- No calcula nada por s铆 solo  solo define la ventana.  
- Debes aplicarle una funci贸n como `.mean()`, `.sum()`, `.max()`, etc.  
- Si no hay suficientes datos para completar la ventana, el resultado ser谩 `NaN`.  

Ejemplo:  
Con `[10, 12, 14, 16]` y `.rolling(3).mean()`   
- Los primeros 2 valores ser谩n `NaN` porque no hay 3 datos a煤n.  
- El tercer valor ser谩 `(10+12+14)/3 = 12.0`.  
- El cuarto valor ser谩 `(12+14+16)/3 = 14.0`.
