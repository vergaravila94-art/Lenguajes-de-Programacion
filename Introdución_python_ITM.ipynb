### Funciones para devolver un valor o operacin:

```python
def cuadrados(x, y):
    return x**2 + y**2 # la funcin finaliza con el return, si se escribe otras lineas despes no se ejecutan

result = cuadrados(7,4)

print(result)
```

Con el resultado se puede invocar nuevamente la funcin y enviandola como argumento

```python
result = cuadrados(result,35)

print(result)
type(result)
```

Las funciones pueden devolver varios resultados y las devuelve como **Tuplas**

En este ejemplo se envia dos argumentos y se invoca la funcion sal_comi, ella tiene dos variables
que devuelven la multiplicacion de x,y y la multiplicacion del resultado de la variable sal
por una costante de 0.30

```python
def sal_comi (x,y):
    sal = x * y
    com = sal * 0.30
    return sal, com

total = sal_comi(10000,80)

print(total)
print(type(total))
```

El resultado de esta tupla se puede dividir y asi manejar dos tuplas diferentes

```python
salario, comision = sal_comi(10000,80)
print(salario)
print(comision)
```

## COUNTER

```python
# Forma avanzada

from collections import counter


ordenes = ["f-1516",
           "f-1516",
           "f-1517",
           "f-1516",
           "f-1516",
           "f-1518",
           "f-1516"
          ]
cantidad_or= Counter(ordenes)

print(cantidad_or)
```
